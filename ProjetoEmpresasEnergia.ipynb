```python
# Import das bibliotecas necessárias para a análise

# Import da biblioteca que irá possibilitar acessar,
# análisar e manipular os dados do dataset.
import pandas as pd

# Import da biblioteca que possui uma grande variedade de gráficos
import seaborn as sns

# Import da biblioteca que possibilita criar e manipular gráficos
import matplotlib.pyplot as plt


```

```python
# Conjunto de dados das ações de empresas do ramo energético na B3
base_dados = pd.read_excel('Dados_Empresas_Energia.xlsx')
```

```python
# Visualizando as 5 primeiras linhas do dataset

print("5 PRIMEIRAS LINHAS DO DATASET")

base_dados.head()
```

```python
# Visualizando as informações gerais do dataset

print("INFORMAÇÕES GERAIS DO DATASET")

base_dados.info()
```

```python
# Série Temporal: setar o indice
# Esse trecho tem como objetivo definir a coluna 'Data' como
# indice do dataframe base_dados.
# Resumindo é um metodo da biblioteca pandas que irá transformar
# uma coluna em indice do Dataframe.
# Agora posso acessar as informações das linhas usando as datas 
# como chave para o indice. 
# O inplace=True ira aplicar a alteração no dataset original
base_dados.set_index('Data', inplace=True)
```

```python
# Verificando se as datas viraram indices

print("VERFICANDO SE AS DATAS VIRARAM INDICES")

base_dados.head()
```

```python
# Antes de iniciar a construção do gráfico, vamos verificar os estilos
# disponiveis do plt.style.use

plt.style.available
```

```python
# Criação do gráfico

# Estilo do gráfico
plt.style.use('seaborn-v0_8-darkgrid')

# Tamanho da imagem do gráfico

plt.figure(figsize=(20, 7))

# titulo do gráfico: Será alinhado a esquerda, tera o tamanho da fonte 18,
# e a largura da fonte será 0.
plt.title('Análise de ações de empresas de energia', loc='left', fontsize=18, fontweight=0)


# Plot do grafico de linhas da petrobras: Terá a color da linha como verde,
# a margem da linha será 4 e a transparencia 0.7

# color: define a cor

# linewidth: Define a largura da linha

# alpha: define a transparência da cor da linha

plt.plot(base_dados.index, base_dados['Petrobras'], color='#008c4a', linewidth=4, alpha=0.7)

# plt.text: Comando do matplotlib que adiciona texto em um gráfico
# base_dados.index[-1]: Acessa o indice do dataframe que no nosso
# caso é a sequência das datas (valores da coluna 'Data')

# -1: seleciona o ultimo valor do indice. Ou seja, você está pegando
# o ultimo indice da série de dados em base_dados.

# base_dados['Petrobras']: Seleciona a coluna chamada 'Petrobras' do DataFrame

# tail(1): pega o ultimo valor dessa coluna. Ou seja, você está acessando
# o valor mais recente da coluna 'Petrobras'. O uso do Tail retorna uma
# série com apenas o último valor, mas como plt.text espera um valor
# escalar, ele pegará apenas esse valor.

# 'Petrobras': Este é o texto que será exibido no gráfico. O nome 'Petrobras'
# será colocado na posição especificada no gráfico.

# color: Defini a cor do texto

# size: Define o tamanho do texto. Definimos o tamanho 'large'

# horizontalalignment: Define o alinhamento horizontal do texto



plt.text(base_dados.index[-1], base_dados['Petrobras'].tail(1), 'Petrobras', color='red', size='large', horizontalalignment='left')

# Rótulo do eixo X
plt.xlabel('Periodo')

# Rótulo do eixo Y
plt.ylabel('Preço do fechamento')
```
